```python
# CONVERTS DICOM TIME INFO TO DATETIME #

def convertToDatetime(date, time):
    """
    Converts the time written in the DICOM tags to datetime format.
    Parameters:
      date - date
      time - time
    Return:
      datetime - date and time in datetime format
    """
    year = int(int(date)/10000)
    month = int(int(date)/100-year*100)
    day = int(int(date)-year*10000-month*100)
    hour = int(float(time)/10000)
    minute = int(float(time)/100-hour*100)
    seconds =  int(float(time)-minute*100-hour*10000)
    return datetime.datetime(year, month, day, hour, minute, seconds)
```

```python
# READS THE DICOM FILE #


def readDicomFiles(filepath):
    """
    Reads the dicom files and returns an array containing the image,
    the dimension of the image, and the ref. dicom file to be able 
    to access dicom tags.
    Parameters:
      filepath - path to the folder containing the dicom files of the image
    Return:
      pixArray - array containing the image
      constPixDim - dimension of the image
      ds - reference dicom file
    """

    # Get the file list
    print "Imported Files :"
    listFilesDCM = []  # creates an empty list
    for dirName, subdirList, fileList in os.walk(filepath):
        for filename in fileList:
            if ".dcm" in filename.lower():  # check whether the file's DICOM
                listFilesDCM.append(os.path.join(dirName,filename))
                print "  > ", filename
    print " "
    
    # Get ref file
    ds = dcm.read_file(listFilesDCM[0])

    # Load dimensions:
    nb_pixel_x = int(ds.Rows)
    nb_pixel_y = int(ds.Columns)
    nb_slices = len(listFilesDCM)
    constPixDim = [nb_slices, nb_pixel_y, nb_pixel_x]
    pixArray = np.zeros(constPixDim, dtype=np.float32)
    
    # loop through all the DICOM files
    for filenameDCM in listFilesDCM:
        dstmp = dcm.read_file(filenameDCM)
        try:
            rescaleIntercept = ds[0x0028,0x1052].value
            rescaleSlope = ds[0x0028,0x1053].value
            pixArray[listFilesDCM.index(filenameDCM),:,:] = dstmp.pixel_array.astype(np.float32) * rescaleSlope + rescaleIntercept
        except KeyError:
            rescaleIntercept = 0
            rescaleSlope = 1

    return pixArray, constPixDim, ds
```

```python
# READS THE DICOM FILE #


def readOneDicomFile(filepath, seriesUID):
    """
    Reads the dicom files and returns an array containing the image
    corresponding to the seriesUID given as an argument, the dimension 
    of the image, and the ref. dicom file to be able to access dicom tags.
    Parameters:
      filepath - path to the folder containing the dicom files of the image
    Return:
      pixArray - array containing the image
      constPixDim - dimension of the image
      ds - reference dicom file
    """

    # Get the file list
    listFilesDCM = []  # creates an empty list
    for dirName, subdirList, fileList in os.walk(filepath):
        for filename in fileList:
            if ".dcm" in filename.lower():  # check whether the file's DICOM
                listFilesDCM.append(os.path.join(dirName,filename))

    # finds ref. img and load dimensions:
    firstSlice = 0
    nboffiles = len(listFilesDCM)
    dstmp = dcm.read_file(listFilesDCM[firstSlice])
    while ((dstmp.SeriesInstanceUID != seriesUID)and(firstSlice<nboffiles-1)):
        firstSlice = firstSlice+1
        dstmp = dcm.read_file(listFilesDCM[firstSlice])
    nb_pixel_x = int(dstmp.Rows)
    nb_pixel_y = int(dstmp.Columns)

    nb_slices = 0
    for i in range(firstSlice,len(listFilesDCM)):
        dstmp = dcm.read_file(listFilesDCM[i])
        if (dstmp.SeriesInstanceUID == seriesUID):
            nb_slices = nb_slices + 1

    constPixDim = [nb_slices, nb_pixel_y, nb_pixel_x]
    pixArray = np.zeros(constPixDim, dtype=np.float32)
    
    for i in range(firstSlice, firstSlice+nb_slices):
        dstmp = dcm.read_file(listFilesDCM[i])
        try:
            rescaleIntercept = dstmp[0x0028,0x1052].value
            rescaleSlope = dstmp[0x0028,0x1053].value
        except KeyError:
            rescaleIntercept = 0
            rescaleSlope = 1
        pixArray[i-firstSlice,:,:] = dstmp.pixel_array.astype(np.float32) * rescaleSlope + rescaleIntercept
    
    ds = dcm.read_file(listFilesDCM[firstSlice])
    
    return pixArray, constPixDim, ds
```

```python
# GET RELEVANT INFORMATION FROM DICOM FILE #

def getAndPrintRelevantInfo(ds, arrayDim):
    """
    Get relevant information from the dicom image.
    Parameters:
      ds - dicom image
      arrayDim - image dimension
    Return:
      pixArray - array containing the image
      constPixDim - dimension of the image
      ds - reference dicom file
    """

    study_date = ds.StudyDate
    scan_time = ds[0x0008,0x0032].value
    acq_datetime = convertToDatetime(study_date,scan_time)

    print "Patient name: ", ds.PatientName
    print "Patient ID: ", ds[0x0020,0x0010].value
    print "File path:",DCM_FILE_PATH
    print " "
    print "Scan date and time: ", acq_datetime
    print "Image size:", arrayDim[2], "x",arrayDim[1], "x", arrayDim[0]
    try:
        print "Rescale slope:", ds[0x0028,0x1052].value
        print "Rescale intercept:", ds[0x0028,0x1053].value
    except:
        print "Rescale slope:", 1
        print "Rescale intercept:", 0
    print " "
    print "System: ", ds[0x0008,0x0070].value
    print "        ", ds[0x0008,0x1010].value

    return acq_datetime
```

```python
# FINDS THE SUV PHANTOM IN THE IMAGE #

# function to find the phantom dimension and center in the image
def findPhantom(img, dimx, dimy, tp):
    """
    Function to find the phantom in the image.
    Parameters:
      img - image array
      dimx - image dimension in x
      dimy - image dimension in y
      tp - threshold to consider to segment out the phantom
    Return:
      centerx - center of the phantom in x direction
      centery - center of the phantom in y direction
      radius - radius of the phantom
    """
    com = ndimage.measurements.center_of_mass(img)

    maxx = ndimage.measurements.maximum(img[com[0],com[1],:])
    x1 = 0
    while (img[com[0],com[1],x1]<maxx*tp):
        x1 = x1+1
    x2 = dimx-1
    while (img[com[0],com[1],x2]<maxx*tp):
        x2 = x2-1
    x1=x1-1
    x2=x2+1
    
    maxy = ndimage.measurements.maximum(img[com[0],:,com[2]])
    y1 = 0
    while (img[com[0],y1,com[2]]<maxy*tp):
        y1 = y1+1
    y2 = dimy-1
    while (img[com[0],y2,com[2]]<maxy*tp):
        y2 = y2-1
    y1=y1-1
    y2=y2+1

    centerx = int(x1+(x2-x1)/2)
    centery = int(y1+(y2-y1)/2)
    
    if ((y2-y1)>(x2-1)):
        radius =(y2-y1)/2
    else:
        radius = (x2-x1)/2

    return [centerx, centery, radius]
```

```python
# FUNCTION TO GET THE STATS INSIDE THE ROI #

def get_stats(img, dimx, dimy, centerx, centery, r, slicenb):
    """
    Returns mean and std dev of the values of the pixels inside the ROI
    Parameters:
      img - image array
      dimx - image dimension in x
      dimy - image dimension in y
      centerx - center of the ROI in x
      centery - center of the ROI in y
      r - radius of the ROI
      slicenb - slice of the image on which are calculated the mean and std. dev.
    Return:
      mean - mean value of the pixels inside the ROI
      std - standard deviation of the value of the pixels inside the ROI
    """
    l = []
    for j in range(0, dimy):
        for i in range(0, dimx):
            if (math.sqrt((i-centerx)*(i-centerx)+(j-centery)*(j-centery))<=r):
                l.append(img[slicenb,j,i])

    return np.mean(l), np.std(l)
```

```python
# FUNCTION TO DISPLAY THE STATS OF THE ROIs #

def dispResults(pxls, dim, cx, cy, r):
    """
    Disp the results:
      pxls - pixel array of the images
      dim - dimension of the pixel array (dim[2]:x dim[1]:y dim[0]: nb of images)
      cx - center of the phantom in x
      cy - center of the phantom in y
      r - radius of the phantom
    Return:
      void
    """
    
    c_mean, c_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy, 0.4*r-2*ROI_MARGIN, SLICE_80kV)
    n_mean, n_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_80kV)
    s_mean, s_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_80kV)
    w_mean, w_stddev = get_stats(pxls, dim[2], dim[1], cx-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_80kV)
    e_mean, e_stddev = get_stats(pxls, dim[2], dim[1], cx+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_80kV)
    
    limitNoiseHU = 0.1 * NOISE_REF_80kV
    if (limitNoiseHU<0.2): limitNoiseHU = 0.2
    
    print 'Le rayon du fantome est', r, 'pixels'
    print 'Le rayon de la ROI centrale est:', int(0.4*r-10), 'pixels'
    print 'Le rayon des ROIs cardinales est:', EXT_ROIS_RADIUS, 'pixels'
    print '   ' 
    print 'Results for the slice nb', SLICE_80kV, '(80kV)'
    print '   ' 
    print '  central circle -> mean:', round(c_mean,2),' & std. dev:', round(c_stddev,2), ' (ref. :', NOISE_REF_80kV, ')' 
    print '  north circle -> mean:', round(n_mean,2),' & std. dev:', round(n_stddev,2)
    print '  south circle -> mean:', round(s_mean,2),' & std. dev:', round(s_stddev,2)
    print '  west circle -> mean:', round(w_mean,2),' & std. dev:', round(w_stddev,2)
    print '  east circle -> mean:', round(e_mean,2),' & std. dev:', round(e_stddev,2)
    if ((abs(c_mean)<4)and(abs(n_mean-c_mean)<4)and(abs(s_mean-c_mean)<4)and(abs(w_mean-c_mean)<4)
        and(abs(e_mean-c_mean)<4)and(abs(c_stddev-NOISE_REF_80kV)<limitNoiseHU)):
        print '  >>> Test succeeded ! '
    else :
        print '  >>> Test Failed...'
    
    c_mean, c_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy, 0.4*r-2*ROI_MARGIN, SLICE_120kV)
    n_mean, n_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_120kV)
    s_mean, s_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_120kV)
    w_mean, w_stddev = get_stats(pxls, dim[2], dim[1], cx-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_120kV)
    e_mean, e_stddev = get_stats(pxls, dim[2], dim[1], cx+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_120kV)
    
    limitNoiseHU = 0.1 * NOISE_REF_120kV
    if (limitNoiseHU<0.2): limitNoiseHU = 0.2
        
    print '   ' 
    print 'Results for the slice nb', SLICE_120kV, '(120kV)'
    print '   ' 
    print '  central circle -> mean:', round(c_mean,2),' & std. dev:', round(c_stddev,2), ' (ref. :', NOISE_REF_120kV, ')' 
    print '  north circle -> mean:', round(n_mean,2),' & std. dev:', round(n_stddev,2)
    print '  south circle -> mean:', round(s_mean,2),' & std. dev:', round(s_stddev,2)
    print '  west circle -> mean:', round(w_mean,2),' & std. dev:', round(w_stddev,2)
    print '  east circle -> mean:', round(e_mean,2),' & std. dev:', round(e_stddev,2)
    if ((abs(c_mean)<4)and(abs(n_mean-c_mean)<4)and(abs(s_mean-c_mean)<4)and(abs(w_mean-c_mean)<4)
        and(abs(e_mean-c_mean)<4)and(abs(c_stddev-NOISE_REF_120kV)<limitNoiseHU)):
        print '  >>> Test succeeded ! '
    else :
        print '  >>> Test Failed...'
    print '   ' 
    print '   ' 
```

```python
# FUNCTION TO REFRESH THE PLOTS #
    
def refreshPlot(minv, maxv, im1, im2, pixArray, fig):    
    """
    Refresh the plot when the sliders are moved:
      minv - min value of the visualization histogram of the image
      maxv - max value of the visualization histogram of the image
      im1 - graph image 80kV
      im2 - graph image 120kV
      pixArray - pixel array of the images
      fig - figure to update
    Return:
      void
    """
    if (minv>maxv):
        print "Erreur: min > max"
    else:
        im1.set_data(pixArray[SLICE_80kV,:,:])
        im1.set_clim(vmin=minv, vmax=maxv)
        im2.set_data(pixArray[SLICE_120kV,:,:])
        im2.set_clim(vmin=minv, vmax=maxv)
    display.clear_output(wait=True)
    display.display(fig)
```

```python
# FUNCTION TO WRITE RESULTS IN EXCEL FILE #
    
def saveToXls(pxls, dim, cx, cy, r, date):    
    """
    Write results in an excel file
      pxls - pixel array of the images
      dim - dimension of the pixel array (dim[2]:x dim[1]:y dim[0]: nb of images)
      cx - center of the phantom in x
      cy - center of the phantom in y
      r - radius of the phantom
      date - date and time of measurements
    Return:
      void
    """
    
    wb = load_workbook(EXCEL_FILE)
    sheet = wb.get_sheet_by_name(EXCEL_SHEET)

    maxrow = 5
    while (sheet['A'+str(maxrow)].value != None):
        maxrow = maxrow+1
    
    sheet['A'+str(maxrow)] = date
    
    # image 80kV
    c_mean, c_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy, 0.4*r-2*ROI_MARGIN, SLICE_80kV)
    n_mean, n_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_80kV)
    s_mean, s_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_80kV)
    w_mean, w_stddev = get_stats(pxls, dim[2], dim[1], cx-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_80kV)
    e_mean, e_stddev = get_stats(pxls, dim[2], dim[1], cx+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_80kV)

    ft = Font(color=colors.RED, bold = True)
    if abs(c_mean)>4:
        sheet['C'+str(maxrow)].font = ft
    sheet['C'+str(maxrow)] = float(c_mean)
    if abs(n_mean-c_mean)>4:
        sheet['D'+str(maxrow)].font = ft
    sheet['D'+str(maxrow)] = float(n_mean)
    if abs(s_mean-c_mean)>4:
        sheet['E'+str(maxrow)].font = ft
    sheet['E'+str(maxrow)] = float(s_mean)
    if abs(e_mean-c_mean)>4:
        sheet['F'+str(maxrow)].font = ft
    sheet['F'+str(maxrow)] = float(e_mean)
    if abs(w_mean-c_mean)>4:
        sheet['G'+str(maxrow)].font = ft
    sheet['G'+str(maxrow)] = float(w_mean)
    if (maxrow>5):
        limitNoiseHU = 0.1 * sheet['H6'].value
        if (limitNoiseHU<0.2): limitNoiseHU = 0.2
        if abs(c_stddev-sheet['H6'].value)>limitNoiseHU:
            sheet['H'+str(maxrow)].font = ft
    sheet['H'+str(maxrow)] = float(c_stddev)
    
    # image 120kV
    c_mean, c_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy, 0.4*r-2*ROI_MARGIN, SLICE_120kV)
    n_mean, n_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_120kV)
    s_mean, s_stddev = get_stats(pxls, dim[2], dim[1], cx, 
                               cy+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, EXT_ROIS_RADIUS, SLICE_120kV)
    w_mean, w_stddev = get_stats(pxls, dim[2], dim[1], cx-0.4*r-EXT_ROIS_RADIUS-ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_120kV)
    e_mean, e_stddev = get_stats(pxls, dim[2], dim[1], cx+0.4*r+EXT_ROIS_RADIUS+ROI_MARGIN, 
                               cy, EXT_ROIS_RADIUS, SLICE_120kV)
    
    if abs(c_mean)>4:
        sheet['J'+str(maxrow)].font = ft
    sheet['J'+str(maxrow)] = float(c_mean)
    if abs(n_mean-c_mean)>4:
        sheet['K'+str(maxrow)].font = ft
    sheet['K'+str(maxrow)] = float(n_mean)
    if abs(s_mean-c_mean)>4:
        sheet['L'+str(maxrow)].font = ft
    sheet['L'+str(maxrow)] = float(s_mean)
    if abs(e_mean-c_mean)>4:
        sheet['M'+str(maxrow)].font = ft
    sheet['M'+str(maxrow)] = float(e_mean)
    if abs(w_mean-c_mean)>4:
        sheet['N'+str(maxrow)].font = ft
    sheet['N'+str(maxrow)] = float(w_mean)
    if maxrow>5:
        limitNoiseHU = 0.1 * sheet['O6'].value
        if (limitNoiseHU<0.2): limitNoiseHU = 0.2
        if abs(c_stddev-sheet['O6'].value)>limitNoiseHU:
            sheet['O'+str(maxrow)].font = ft
    sheet['O'+str(maxrow)] = float(c_stddev)

    wb.save(EXCEL_FILE)
```
